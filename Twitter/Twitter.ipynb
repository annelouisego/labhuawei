```python
import GetOldTweets3 as got
```

```python
tweetCriteria = got.manager.TweetCriteria().setQuerySearch('huaweiandroid OR HuaweiAndroid OR huaweiAndroid OR Huaweiandroid OR HUAWEIANDROID OR huaweiandroidban OR huawei android OR HuaweiBan OR huaweigoogleban OR huaweigoogle OR huaweitrump OR HuaweyTrump')\
                                           .setLang("en")\
                                           .setSince("2019-05-14")\
                                           .setUntil("2019-05-16")\
                                           .setMaxTweets(3000)
tweets = got.manager.TweetManager.getTweets(tweetCriteria)
print(tweets.text)
```

```python
import pprint
pprint.pprint(tweets[0].text)

len(tweets)
```

```python
for t in tweets:
    print(t.text + " - " + t.id)
    print("--------------------------------------------")
```

```python
import pandas as pd
```

```python
dict_ = {'id': [], 'date': [], 'text': [], 'hashtags': []}
for t in tweets:
    dict_['id'].append(t.id)
    dict_['date'].append(t.date)
    dict_['text'].append(t.text)
    dict_['hashtags'].append(t.hashtags)
    
print(dict_) 
```

```python
df = pd.DataFrame.from_dict(dict_, orient="columns")
df = df.set_index("id")
df.sort_values(by='date', inplace=True, ascending=False)
df.head(100)
```
